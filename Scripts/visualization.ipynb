```python
import csv
import pandas as pd
import altair as alt
from itertools import combinations
```

## RQ1

```python
rq1_data = pd.read_csv("../data/SQL_outputs/RQ1.csv")
```

```python
imdb_hist = alt.Chart(rq1_data).mark_bar(opacity=0.5).encode(
    alt.X("IMDBRATINGS:Q", bin=alt.Bin(maxbins=20), title="IMDb Rating"),
    alt.Y("count()", stack=None, title="Number of Films"),
    alt.Color("ADULT_FLAG:N", title="Film Category",
             scale=alt.Scale(
            domain=["Adult", "Non-Adult"],
            range=["red", "green"]
        )),
    alt.Order("ADULT_FLAG:N")
).properties(
 #   title="IMDb Rating Distribution: Adult vs Non-Adult Films (2015–2020)",
    width=300
)

rt_hist = alt.Chart(rq1_data).mark_bar(opacity=0.5).encode(
    alt.X("ROTTENTOMATOESRATINGS:Q", bin=alt.Bin(maxbins=20), title="Rotten Tomatoes Rating"),
    alt.Y("count()", stack=None, title="Number of Films"),
    alt.Color("ADULT_FLAG:N", title="Film Category"),
    alt.Order("ADULT_FLAG:N")
).properties(
  #  title="Rotten Tomatoes Rating Distribution: Adult vs Non-Adult Films (2015–2020)",
    width=300
)
rq1_charts = (imdb_hist | rt_hist).properties(
    title="IMDb vs Rotten Tomatoes Rating Distribution: Adult vs Non-Adult Films (2015–2020)"
).configure_axis(
    labelFontSize=15,
    titleFontSize=18
).configure_title(
    fontSize=22
).configure_legend(
    labelFontSize=15,
    titleFontSize=18
)
rq1_charts
```

```python
rq1_charts.save('../figures/rq1_charts.png')
```

```python
rq2_data1 = pd.read_csv("../data/SQL_outputs/RQ2_1.csv")
```

```python
rq2_chart_single_genre = alt.Chart(rq2_data1).mark_bar(opacity=0.5).encode(
    x=alt.X("GENRE:N", sort="-y", title="Genre", axis=alt.Axis(labelAngle=-30)),
    y=alt.Y("GENRE_COUNT:Q", title="Number of Genre Appearances"),
    color=alt.Color("ADULT_FLAG:N", title="Film Category",
                   scale=alt.Scale(
            domain=["Adult", "Non-Adult"],
            range=["red", "green"]
        )),
    xOffset="ADULT_FLAG:N"
).properties(
    title="Genre Appearances in Adult vs Non-Adult Films (2015–2020)",
    width=800
).configure_axis(
    labelFontSize=15,
    titleFontSize=18
).configure_title(
    fontSize=22
).configure_legend(
    labelFontSize=15,
    titleFontSize=18
)
rq2_chart_single_genre
```

### 2-2

```python
rq2_data2 = pd.read_csv("../data/SQL_outputs/RQ2_2.csv")
```

```python
adult_top10 = (
    rq2_data2[rq2_data2["ADULT_FLAG"] == "Adult"]
    .value_counts("GENRES")
    .reset_index(name="count")
    .head(10)
)
nonadult_top10 = (
    rq2_data2[rq2_data2["ADULT_FLAG"] == "Non-Adult"]
    .value_counts("GENRES")
    .reset_index(name="count")
    .head(10)
)
```

```python
adult_chart_combo = alt.Chart(adult_top10).mark_bar(opacity=0.5, color="red").encode(
    x=alt.X("count:Q", title="Number of Films"),
    y=alt.Y("GENRES:N", sort="-x", title="Genre Combination")
).properties(
    title="Top 10 Genre Combinations in Adult Films",
    width=350,
    height=300
)
nonadult_chart_combo = alt.Chart(nonadult_top10).mark_bar(opacity=0.5, color="green").encode(
    x=alt.X("count:Q", title="Number of Films"),
    y=alt.Y("GENRES:N", sort="-x", title="Genre Combination")
).properties(
    title="Top 10 Genre Combinations in Non-Adult Films",
    width=350,
    height=300
)

rq2_charts_combo = (adult_chart_combo | nonadult_chart_combo).configure_axis(
    labelFontSize=15,
    titleFontSize=18
).configure_title(
    fontSize=22
)
rq2_charts_combo
```

```python
rq2_charts_combo.save('../figures/rq2_chart2.png')
```

## RQ3

```python
rq3_data = pd.read_csv("../data/SQL_outputs/RQ3.csv")
```

```python
rq3_scatter = alt.Chart(rq3_data).mark_circle(size=70, opacity=0.5).encode(
    x=alt.X("IMDBRATINGS:Q", title="IMDb Rating"),
    y=alt.Y("GROSS:Q", 
            title="International Box Office Revenue",),
    color=alt.Color("ADULT_FLAG:N", title="Film Category",
                   scale=alt.Scale(
            domain=["Adult", "Non-Adult"],
            range=["red", "green"]
        ),),
    tooltip=["TITLE","IMDBRATINGS","GROSS","ADULT_FLAG"]
).properties(
    title="International Box Office vs IMDb Rating (2015–2020)",
    width=500,
    height=350
).interactive(bind_x=True, bind_y=True)

rq3_reg = rq3_scatter + alt.Chart(rq3_data).transform_regression(
    "IMDBRATINGS",
    "GROSS",
    groupby=["ADULT_FLAG"]
).mark_line().encode(
    x="IMDBRATINGS:Q",
    y="GROSS:Q",
    color="ADULT_FLAG:N"
)

rq3_chart=rq3_reg.configure_axis(
    labelFontSize=15,
    titleFontSize=18
).configure_title(
    fontSize=22
).configure_legend(
    labelFontSize=15,
    titleFontSize=18
)
rq3_chart
```

```python
rq3_reg.save('../figures/rq3_chart.png')
```
